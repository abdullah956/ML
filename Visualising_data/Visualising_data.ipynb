```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("WranglingData.csv")
data.head(5)
```

```python
plt.figure()
plt.scatter(x=data['Quantity Purchased'],y=data['Price'])
```

```python
sns.scatterplot(data=data ,x='Quantity Purchased',y='Price',hue='User ID',size='Product ID')
# You can specify a categorical column for color hue
# You can specify a numerical column for point size
```

```python
data.plot(kind="scatter",x='Quantity Purchased',y='Price')
```

```python
data.Price.plot(kind='hist')
```

```python
sns.displot(data['Price'])
```

```python
sns.boxplot(x=data['Price'])
plt.show()
```

```python
fig, ax = plt.subplots()
# figure and axes 
ax.set(title="Graph title", xlabel="Label of x axis (units)", 
ylabel="Label of y axis (units)")
ax.plot()
```

```python
#plot legend
fig, ax = plt.subplots()
data.groupby('Product ID')['User ID'].mean().plot(ax=ax, 
label='Price')
ax.legend()
```

```python
#alternate method
fig, ax = plt.subplots()
data.groupby('Product ID')['User ID'].mean().plot(ax=ax)
ax.legend(['Price'])
```

```python
data.groupby('Product ID')['User ID'].mean().plot(ls='-.', color='r', lw=3)
```

```python
data.plot(kind='scatter', x='Product ID', y='User ID', figsize=(20,10))
plt.savefig('horsepower_weight_scatter.png')
```
