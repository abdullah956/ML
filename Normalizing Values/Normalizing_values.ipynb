```python
data['date'] = pd.to_datetime(data['date']) # Typecasting changing datatype of date column
data.dtypes
```

```python
#Transformation
#map () method
def expansion_user_type(u_type):
    if u_type in ['a','b']:
        return 'New'
    elif u_type =='c':
        return 'Existing'
    elif u_type =='d':
        return 'Royal-Existing'
    else:
        return 'error'
```

```python
data['user_class'] = data['user_type'].map(expansion_user_type)
```

```python
data[['user_type','user_class']]
```

```python
#applymap() method
data['purchased_week'] = data[['date']].applymap(lambda dt:dt.week if not pd.isnull(dt.week) else 0)
data['purchased_week']
```

```python
#apply method
import numpy as np
data.select_dtypes(include=np.number).apply(lambda x: x.max()-x.min())
```

```python
df_dropped = data.dropna(subset=['date'])
print("After Dropping rows with missing values :")
print(df_dropped)
print("Rows and Columns : " ,df_dropped.shape)
```

```python
df_dropped = data.dropna(subset=['price'])
print("After Dropping rows with missing values :")
print(df_dropped)
print("Rows and Columns : " ,df_dropped.shape)
```
