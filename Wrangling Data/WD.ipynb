```python
data['date'] = pd.to_datetime(data['date'])
data.dtypes
```

```python
def expansion_user_type(u_type):
    if u_type in ['a','b']:
        return 'New'
    elif u_type =='c':
        return 'Existing'
    elif u_type =='d':
        return 'Royal-Existing'
    else:
        return 'error'
```

```python
data['user_class'] = data['user_type'].map(expansion_user_type)
```

```python
data[['user_type','user_class']]
```

```python
data['purchased_week'] = data[['date']].applymap(lambda dt:dt.week if not pd.isnull(dt.week) else 0)
data[data['purchased_week']<10]
#data['purchased_week']
```
