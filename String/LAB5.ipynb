```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("dataset.csv")
```

```python
dataset.head(5)
```

```python
dataset["Tweets"]=dataset["Tweets"].apply(lambda x: x.lower())
dataset["Tweets"]
```

```python
dataset["Tweets"]=dataset["Tweets"].str.replace("[^a-zA-Z0-9]", " ", regex=True)
dataset["Tweets"]
```

```python
dataset["Tweets"]=dataset["Tweets"].str.replace("[^a-zA-Z]", " ", regex=True)
dataset["Tweets"]
```

```python
dataset["Tweets"]=dataset["Tweets"].apply(lambda x: x.split())
dataset["Tweets"]
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
wordsEng = stopwords.words('english')
dataset["Tweets"]=dataset["Tweets"].apply(lambda x:[item for item in x if item not in wordsEng])
dataset["Tweets"]
```

```python
dataset["Tweets"]=dataset["Tweets"].apply(lambda x:[x1 for x1 in x if not any(c.isdigit() for c in x1)])
dataset["Tweets"]
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
dataset["Tweets"]=dataset["Tweets"].apply(lambda x: [ps.stem(y) for y in x])
dataset["Tweets"]
```

```python
dataset["Tweets"]=dataset["Tweets"].apply(lambda x: [w for w in x if len(w)>3])
dataset["Tweets"]
```

```python
dataset["Tweets"]=dataset["Tweets"].apply(lambda x: " ".join(x))
dataset["Tweets"]
```
