```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
data = pd.read_csv("Iris.csv")
data["Species"].head(10)
```

```python
Species = np.unique(data["Species"])
Species
```

```python
gle = LabelEncoder()
Species_Labels = gle.fit_transform(data["Species"])
Species_Mapping = { i : label  for i , label in enumerate(gle.classes_)}
Species_Mapping
```

```python
data["Species_Label"] = Species_Labels
data.head(10)
```
