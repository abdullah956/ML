```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv("Pokemon.csv")
data.head()
```

```python
data[['HP','Attack','Defense']].head(5)
```

```python
data_atack_def = data[['Attack','Defense']].head(5)
```

```python
pf = PolynomialFeatures(degree=2 , interaction_only=False , include_bias=False)
result = pf.fit_transform(data_atack_def)
result
```

```python
pd.DataFrame(pf.powers_ , columns=['Attack','Defense'])
```

```python
ntr_features = pd.DataFrame(result, columns=['Attack','Defense','Attack^2','Attack x Defense','Defense^2'])
ntr_features
```
