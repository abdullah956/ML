```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('fcc_2016_coder_survey_subset.csv')
data.head(10)
```

```python
data[['ID.x','EmploymentField','Age','Income']].head(10)
```

```python
fig , ax = plt.subplots()
data['Age'].hist(color='red')
ax.set_title('Developer Age Histogram',fontsize=12)
ax.set_xlabel('Age',fontsize=12)
ax.set_ylabel('Frequency',fontsize=12)
```

```python
data['Age_bin_round'] = np.array(np.floor(np.array(data['Age'])/10))
data[['Age','Age_bin_round']].head(10)
```

```python
data[['Age','Age_bin_round']].iloc[1071:1076]
```

```python
bin_ranges = [0,15,30,45,60,75,100]
bin_names = [1,2,3,4,5,6]
data['bin_custom_range'] = pd.cut(np.array(data['Age']),bins=bin_ranges)
data['bin_custom_label'] = pd.cut(np.array(data['Age']),bins=bin_ranges, labels=bin_names)
data[['Age','Age_bin_round','bin_custom_range','bin_custom_label']].iloc[1071:1076]
```
